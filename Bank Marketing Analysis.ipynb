### Handling categorical values

```python
first_condition = (bank_df["job"].value_counts())/len(bank_df)
second_condition = (bank_df["marital"].value_counts())/len(bank_df)
third_condition = (bank_df["education"].value_counts())/len(bank_df)
fourth_condition = (bank_df["contact"].value_counts())/len(bank_df)
fifth_condition = (bank_df["month"].value_counts())/len(bank_df)
sixth_condition = (bank_df["poutcome"].value_counts())/len(bank_df)

bank_df["job_encode"] = bank_df["job"].apply(lambda a : first_condition[a])
bank_df["marital_encode"] = bank_df["marital"].apply(lambda b : second_condition[b])
bank_df["education_encode"] = bank_df["education"].apply(lambda c : third_condition[c])
bank_df["contact_encode"] = bank_df["contact"].apply(lambda d : fourth_condition[d])
bank_df["month_encode"] = bank_df["month"].apply(lambda e : fifth_condition[e])
bank_df["poutcome_encode"] = bank_df["poutcome"].apply(lambda f : sixth_condition[f])
```

```python
bank_df.replace({"yes":1,"no":0},inplace=True)
```

```python
bank_df.drop(columns=["job","marital","education","contact","month","poutcome"],inplace=True,axis=1)
bank_df.head()
```

```python
bank_df.shape
```

#### Insights:

The categorical columns have been handled.

### Handling numerical values

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
bank_df[["age","balance","day","duration","campaign","pdays","previous"]] = scaler.fit_transform(bank_df[["age","balance","day","duration","campaign","pdays","previous"]])
bank_df
```

#### Insights:

The numerical columns have been handled using **StandardScaler** as the columns are skewed.
